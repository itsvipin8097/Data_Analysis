```python
df.head()
```

```python
sns.histplot(df["quality"],bins=range(3, 9), kde=False, color="green")
plt.title('Histogram of Wine Quality')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(df[['alcohol', 'pH', 'citric acid', 'quality']])
plt.suptitle('Pair Plots of Selected Attributes', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='quality', data=df, palette='pastel')
plt.title('Distribution of Wine Types')
plt.xlabel('Wine Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
for attr in df:
    sns.violinplot(x='quality', y=attr, data=df, inner="box")
    plt.title(f'Violin Plot of {attr} by Wine Quality')
    plt.show()
```

```python
plt.figure(figsize=(8, 4))
df['quality'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'lightgreen'])
plt.title('Distribution of Wine Types')
plt.show()
```
