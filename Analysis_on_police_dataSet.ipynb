```python
Violation_counts={"Speeding":957,"Moving violation":895,"Equipment":704,"Registration/plates":334,"Other":210,"Seat belt":96}
```

```python
df_violations = pd.DataFrame(list(Violation_counts.items()), columns=['Violation Type', 'Count'])
```

```python
df_violations.head()
```

```python
import matplotlib.pyplot as plt

# Assuming df_violations is your DataFrame
plt.figure(figsize=(10, 6))

# Loop through each row in the DataFrame and plot a bar for each violation type
for index, row in df_violations.iterrows():
    plt.bar(row['Violation Type'], row['Count'], label=row['Violation Type'])

# Set labels and title
plt.xlabel('Violation Type')
plt.ylabel('Count')
plt.title('Violations Count by Type')

# Display the legend
plt.legend()

# Show the plot
plt.show()
```
